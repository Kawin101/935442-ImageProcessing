**2.Image fundamental**

*   RGB 2 CMYK

```python
# RGB 2 CMYK
r = 90
g = 120
b = 150
# rc = r/255
# gc = g/255
# bc = b/255
rc = r/255
gc = g/255
bc = b/255
K = 1-max(rc,gc,bc)
C = ((1-rc-K)/(1-K))*100
M = ((1-gc-K)/(1-K))*100
Y = ((1-bc-K)/(1-K))*100
K = K*100
print(int(C),int(M),int(Y),int(K))
```

```python
# CMYK 2 RGB
C = 100
M = 40
Y = 80
K = 20
r = (1-(C/100))*(1-(K/100))
g = (1-(M/100))*(1-(K/100))
b = (1-(Y/100))*(1-(K/100))
r = r*255
g = g*255
b = b*255
print(int(r),int(g),int(b))
```

**3.2 Image filtering**


*   Average filtering




```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Average filtering
img = cv2.imread("/content/drive/MyDrive/รูปภาพและวิดีโอ/studentphoto.jfif",0)
f = np.ones((5,5), np.float32)/25
avg = cv2.filter2D(img, -1, f)
plt.imshow(img, cmap = "gray")
plt.show()
plt.imshow(avg, cmap = "gray")
plt.show()
```



*   Median filtering


```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Median filtering
img = cv2.imread("/content/drive/MyDrive/รูปภาพและวิดีโอ/studentphoto.jfif",0)
median = cv2.medianBlur(img, 5) # แนะนำให้ใส่เลขคี่
plt.subplot(121), plt.imshow(img, cmap = "gray")
plt.subplot(122), plt.imshow(median, cmap = "gray")
plt.show()
```

*   Noise filtering

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
from skimage.util import random_noise

img = cv2.imread("/content/drive/MyDrive/รูปภาพและวิดีโอ/studentphoto.jfif",0)
noise = random_noise(img, mode="s&p", amount = 0.3)
f = np.ones((5,5), np.float32)/25
avg = cv2.filter2D(noise, -1, f)
median = cv2.medianBlur(noise.astype("float32"),5)
plt.subplot(221), plt.imshow(img, cmap = "gray")
plt.subplot(222), plt.imshow(noise, cmap = "gray")
plt.subplot(223), plt.imshow(avg, cmap = "gray")
plt.subplot(224), plt.imshow(median, cmap = "gray")
plt.show()
```
