# 3.3 Image filtering: Image sharpening

```python
# Image sharpening
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('Cat.jpg') # BGR
gray = cv2.imread('Cat.jpg', 0) # gray
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# filterAVG = np.ones((7,7), np.float32)/49
# img_smoothing = cv2.filter2D(gray, -1, filterAVG)
# img_smoothing = cv2.medianBlur(gray,7)
filterSharp = np.array([[-1,-1,-1],[-1,9,-1],[-1,-1,-1]])
img_sharpen = cv2.filter2D(gray, -1, filterSharp)
print(filterSharp)
plt.subplot(121), plt.imshow(gray, cmap = 'gray')
plt.subplot(122), plt.imshow(img_sharpen, cmap = 'gray')
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img = cv2.imread('Cat.jpg') # BGR
gray = cv2.imread('Cat.jpg', 0) # gray
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

Th = 30
_, obj = cv2.threshold(gray,Th,255,cv2.THRESH_BINARY)
st = cv2.getStructuringElement(cv2.MORPH_RECT,(5,5))
obj = cv2.dilate(obj, st, iterations = 1)
regions, hierarchy = cv2.findContours(obj, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
for r in regions:
  (x,y,w,h) = cv2.boundingRect(r)
  if (w*h) > 50000:
    cv2.rectangle(imgRGB, (x,y), (x+w,y+h), (255,255,0), 2)
plt.imshow(imgRGB, cmap='gray')
plt.show()
```
