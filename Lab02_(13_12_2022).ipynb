```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

gray = cv2.imread('/content/drive/MyDrive/รูปภาพและวิดีโอ/brown-hairstreak-gde1fcb264_1920.jpg',0)
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
eq = cv2.equalizeHist(gray)
histeq = cv2.calcHist([eq], [0], None, [256], [0, 256])

plt.subplot(221), plt.imshow(gray, cmap='gray')
plt.subplot(222), plt.hist(gray, bins=10)
plt.subplot(223), plt.imshow(eq, cmap='gray')
plt.subplot(224), plt.hist(histeq, bins=10)
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

gray = cv2.imread('/content/drive/MyDrive/รูปภาพและวิดีโอ/brown-hairstreak-gde1fcb264_1920.jpg',0)
hist = cv2.calcHist([gray], [0], None, [256], [0, 256])
eq = cv2.equalizeHist(gray)
histeq = cv2.calcHist([eq], [0], None, [256], [0, 256])
para = cv2.createCLAHE(clipLimit=2.0, tileGridSize=(8,8))
aeq = para.apply(gray) # a = adaptive (แบ่งส่วนภาพ)

plt.subplot(321), plt.imshow(gray, cmap='gray')
plt.subplot(322), plt.hist(gray, bins=10)
plt.subplot(323), plt.imshow(eq, cmap='gray')
plt.subplot(324), plt.hist(eq, bins=10)
plt.subplot(325), plt.imshow(aeq, cmap='gray')
plt.subplot(326), plt.hist(aeq, bins=10)
plt.show()
```
